# with list



```python
inputs=weights=list(range(1,11))
print("inputs ",inputs)
print("outputs ",weights)
bias=5
output=0

for i in range(len(inputs)):
    output=output+inputs[i]*weights[i]

output=output+bias

print(f"output:{output},bias:{bias}")
```
