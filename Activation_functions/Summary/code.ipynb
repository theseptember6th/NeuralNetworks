```python
import matplotlib.pyplot as plt  # For plotting graphs
import numpy as np               # For exp, log etc.,
```

### Step Activation


Screenshot 2022-01-19 081348.png0_L3sNJaTdfd4B5ZjF.png

```python
def step(x):
  if x>=0:
    return 1
  else:
    return 0
```

```python
x=np.arange(-6,6,0.01)
step_output = [step(i) for i in x]

# Plotting
fig, ax = plt.subplots(figsize=(9, 5))
ax.plot(x,step_output, color="#307EC7", linewidth=3, label="step")
ax.legend(loc="upper left", frameon=False)
fig.show()
```

### Sigmoid Function



Screenshot 2022-01-19 153224.png3-s2.0-B9780128154588000074-f07-02-9780128154588.jpg

```python
def sigmoid(x):
  s = (1/(1+np.exp(-x)))   # sigmoid function
  return s
```

```python
x=np.arange(-6,6,0.01)
# Plotting
fig, ax = plt.subplots(figsize=(9, 5))
ax.plot(x,sigmoid(x), color="#307EC7", linewidth=3, label="sigmoid")
ax.legend(loc="upper left", frameon=False)
fig.show()
```

### Tanh Activation

Screen_Shot_2020-05-27_at_4.23.22_PM_dcuMBJl.png

```python
def tanh(x):
    t=(np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x))
    # t = np.tanh(x)
    return t
```

```python
x=np.arange(-6,6,0.01)
# Plotting
fig, ax = plt.subplots(figsize=(9, 5))
ax.plot(x,tanh(x), color="#307EC7", linewidth=3, label="tanh")
ax.legend(loc="upper left", frameon=False)
fig.show()
```

### ReLU Activation

main-qimg-621e5e8adfc3468cf9dcabe1d849dcc8.png

```python
def relu(x):
  y = max(0,x)
  return y
```

```python
x=np.arange(-6,6,0.01)
relu_output = [relu(i) for i in x]
# Plotting
fig, ax = plt.subplots(figsize=(9, 5))
ax.plot(x,relu_output, color="#307EC7", linewidth=3, label="relu")
ax.legend(loc="upper left", frameon=False)
fig.show()
```

### Leaky ReLU

Screenshot 2022-02-02 201903.pngScreenshot 2022-02-01 092015.png

```python
def leaky_relu(x):
  y = max(0.1*x,x)
  return y
```

```python
x=np.arange(-6,6,0.01)
leaky_relu_output = [leaky_relu(i) for i in x]
# Plotting
fig, ax = plt.subplots(figsize=(9, 5))
ax.plot(x,leaky_relu_output, color="#307EC7", linewidth=3, label="relu")
ax.legend(loc="upper left", frameon=False)
fig.show()
```

### Softplus Activation

1_invMNMhGkhz2bwvdcoJXPA.png

```python
def softplus(x):
  s = np.log(1+np.exp(x))
  return s
```

```python
x=np.arange(-6,6,0.01)
# Plotting
fig, ax = plt.subplots(figsize=(9, 5))
ax.plot(x,softplus(x), color="#307EC7", linewidth=3, label="softplus")
ax.legend(loc="upper left", frameon=False)
fig.show()
```

### GeLU Activation

Screenshot 2022-02-23 210836.png

```python
def sigmoid(x):
  s = (1/(1+np.exp(-x)))   # sigmoid function
  return s

def gelu(x):
  g = x * sigmoid(1.702*x)
  return g
```

```python
z=np.arange(-6,6,0.01)

fig, ax = plt.subplots(figsize=(9, 5))
ax.plot(z,gelu(z), color="#307EC7", linewidth=3, label="GeLU")
ax.legend(loc="upper left", frameon=False)
fig.show()
```

### Swish Activation

60be24d08129ce48ae90858e_math-20210607 (16).png1_SZ19lV6FT24KtdZsvCOCaA.png

```python
def swish(x):
  s = x* (1/(1+np.exp(-x)))
  return s
```

```python
z=np.arange(-6,6,0.01)

fig, ax = plt.subplots(figsize=(9, 5))
ax.plot(z,swish(z), color="#307EC7", linewidth=3, label="swish")
ax.legend(loc="upper left", frameon=False)
fig.show()
```

### Mish Activation

Screenshot 2022-02-25 185724.png1_RuNM4YV8ZuitdLgkq_MPUw.png

```python
def mish(x):
  m = x * np.tanh(np.log(1+np.exp(x)))
  return m
```

```python
z=np.arange(-6,6,0.01)

fig, ax = plt.subplots(figsize=(9, 5))
ax.plot(z,mish(z), color="#9621E2", linewidth=3, label="mish")
ax.legend(loc="upper left", frameon=False)
fig.show()
```

### Softmax Activation

1_ReYpdIZ3ZSAPb2W8cJpkBg.jpeg

```python
def softmax(x):
    """Compute softmax values for each sets of scores in x."""
    e_x = np.exp(x)
    return e_x / e_x.sum(axis=0) 
```

```python
x = [1.3,5.1,2.2,0.7,1.1]
print(softmax(x))
print(softmax(x).sum())
```
